# Exploratory Data Analysis Report
## MovieLens 32M Dataset

---

This report presents key findings from the data cleaning and exploratory data analysis of the MovieLens 32M dataset (ml-32m). The dataset was filtered and analyzed to support a movie recommendation system.

**Key Findings:**
- **200,947 users** and **13,026 movies** retained after cleaning
- **Drama** dominates both movie count and total rating interactions across all genres
- **Film-Noir** and **War** achieve the highest average ratings despite lower popularity
- **Weak positive correlation (0.24)** between popularity and average rating — popularity is not a reliable source for quality
- **1995-era movies** show a bias in rating counts due to longer accumulation periods


---

## 1. Dataset Overview

### 1.1 Dataset Description

- **Source:** MovieLens 32M (ml-32m), GroupLens Research
- **Original Size:** 32 million ratings
- **Files:** ratings.csv, movies.csv, tags.csv, links.csv
- **Retained after cleaning:** 27.7M+ ratings across 13,026 movies and 200,947 users

### 1.2 Feature Overview

**ratings.csv:**
- userId — unique user identifier
- movieId — unique movie identifier
- rating — user rating (0.5 to 5.0, in 0.5 increments)
- timestamp — Unix timestamp of the rating (converted to datetime)

**movies.csv:**
- movieId — unique movie identifier
- title — movie title with release year
- genres — pipe-separated list of genres

**tags.csv:**
- userId, movieId, tag, timestamp — user-generated tags per movie

**links.csv:**
- movieId, imdbId, tmdbId — cross-references to external databases

---

## 2. Data Cleaning Summary

### 2.1 Missing Values

- **ratings.csv:** No missing values
- **movies.csv:** No missing values
- **tags.csv:** 17 missing tag values → rows dropped
- **links.csv:** 124 missing tmdbId values → rows dropped

### 2.2 Duplicates

- No duplicate rows found across any of the four DataFrames

### 2.3 Data Types

- All columns held correct data types with no casting required
- Timestamps in ratings and tags were converted from Unix integer to datetime format

### 2.4 Rating Range Validation

- Ratings confirmed to fall within the valid range of 0.5 to 5.0
- Mean rating: 3.54 | Std: 1.06


---

## 3. Dataset Scale

### 3.1 Users & Movies

- **Total Users:** 200,947
- **Total Movies:** 13,026

### 3.2 Most Rated Movies

- The top-rated movies by rating count are heavily skewed toward **1995-era titles**
- This is a bias effect — older movies have had longer to accumulate ratings, not necessarily because they are better
- This reinforces the finding that rating count alone is not a reliable quality signal

---

## 4. Genre Analysis

### 4.1 Genre Diversity

- The dataset contains **20 unique genres** (including a '(no genres listed)' category)
- Movies are frequently multi-genre, so genre counts exceed movie counts when exploded

### 4.2 Number of Movies per Genre

- **Drama** has the highest number of movies in the catalog, followed by **Comedy** and **Thriller**
- Less popular genres such as **Film-Noir**, **IMAX**, and **Western** have comparatively few titles

### 4.3 Number of Ratings per Genre

- **Drama** accumulates the most user-rating interactions by a significant margin
- This is consistent with Drama having the largest movie count — more titles means more area for ratings
- **Comedy** and **Action** follow in total interaction volume

### 4.4 Average Rating per Genre

- **Film-Noir** achieves the highest average rating (3.93), followed by **War** (3.85) and **Crime** (3.78)
- **Horror** and **Comedy** receive the lowest average ratings (3.56 and 3.59 respectively)
- The range across all genres is relatively narrow (3.56 – 3.93), suggesting genre alone is not a strong predictor of rating

---

## 5. Correlation Analysis

### 5.1 Movie Popularity vs. Average Rating

We explored whether more popular movies (higher rating count) tend to receive better average ratings.

**Finding: Weak positive correlation (r = 0.24)**

| | mean | count |
|---|---|---|
| mean | 1.00 | 0.24 |
| count | 0.24 | 1.00 |

**Interpretation:**
- A slight tendency exists for more-rated movies to score higher, but the relationship is weak
- Popularity is driven by factors like age, marketing, and cultural relevance — not purely quality
